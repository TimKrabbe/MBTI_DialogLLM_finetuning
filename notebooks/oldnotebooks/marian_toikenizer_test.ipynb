# Testing Placeholder Strategies for MarianMT Tokenizer
We want to find a way to keep the Modell from translating a masking token during backtranslation, so that we don't add any noise in the process. 
The idea here is to try different placeholders for the mask.

```python
from transformers import MarianTokenizer

tok_en_de = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-de")
tok_de_en = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-de-en")
tok = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-de")

```

```python
candidates = ["STX", "PHX", "MBT", "XMB", "ZZZ"]
for word in candidates:
    print(f"{word} en-de: {tok_en_de.tokenize(word)}")
    print(f"{word} de-en: {tok_de_en.tokenize(word)}")
```

```python
# Teste ob ein Wort als single token tokenisiert wird
candidates = ["John", "Peter", "Anna", "Max"]
for word in candidates:
    print(f"{word} en-de: {tok_en_de.tokenize(word)}")
    print(f"{word} de-de: {tok_de_en.tokenize(word)}")

```
